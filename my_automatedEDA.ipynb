```python
def load_data():
    path = input('Please Enter path of data or file name')

    try:

        if path.endswith(".pdf"):
            with open(path,'rb') as pdf_file:
                data = PyPDF2.PdfFileReader(pdf_file)
                return data.read()
                #print(data.read())
        elif path.endswith(".doc" or ".docx"):
            with open(path,'rb') as document_file:
                data= textract.process(document_file)
                return  data.read()
                #print(data.read())
        elif path.endswith(".xlsx"):
            with open(path,'rb') as excel_file:
                data = pd.read_excel(excel_file)
               # print(data.head())
                return data
                
        elif path.endswith(".csv"):
            with open(path,'rb') as csv_file:
                data=pd.read_csv(csv_file)
                #print(data.head())
                return data
                
        elif path.endswith(".json"):
            with open(path,'rb') as json_file:
                data = pd.read_json(json_file)
                #print(data.to_string())
                return data.to_string()
                
        elif path.endswith(".txt"):
            data=open(path,'rb')
            return data.read()
    except:
        print(f"Could not read file {path}")
```

```python
def data_pereparation(data):
    data.columns = data.columns.str.lower()
    print(data.columns)
    print(data.info())
    
```

```python
def heatmap(data):
    print('Heatmap correlation ')
    plt.figure(figsize=((15,6)))
    sns.heatmap(data.corr(),annot=True)
    plt.show()
```

```python
def scatter_plot(data):
    print('Scatter Plot')
    print(data.columns)
    x_column = input("Enter X-axis column : ")
    y_column =input("Enter Y-axis column : ")
    plt.figure(figsize=((15,6)))
    sns.scatterplot(data[x_column],data[y_column])
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()
```

```python
def histogram_plot(data):
    print('Histogram plot : ')
    print(data.columns)
    selected_column_histogram = input("Enter a column for the histogram : ")
    plt.figure(figsize=(8, 6))
    values, bins, patches = plt.hist(selected_column_histogram, bins=10, edgecolor="white", linewidth=3)
    for i in range(0, 3):
        patches[i].set_facecolor('g')
    for i in range(3, 6):
        patches[i].set_facecolor('r')
    for i in range(6, len(patches)):
        patches[i].set_facecolor('black')
    plt.grid(True)
    plt.show()
```

```python
def pie_chart (data):
    print(" Pie Chart")
    print(data.columns)
    plt.figure(figsize=(6, 6))
    d = data.head(1000)
    counts = d[input("Enter a column for the Pie chart : ")].value_counts()
    plt.pie(counts, labels=counts.index.values.tolist())
    plt.axis('equal')
    plt.show()
```

```python
import pandas as pd
import PyPDF2
import textract
import matplotlib.pyplot as plt
import seaborn as sns
data=load_data()
data_pereparation(data)
```

```python
while(True):
    a=int(input("please Enter  : 1 to apply heatmap "
                "       Enter  : 2 to apply scatterplot"
                "       Enter  : 3 to apply histogram"
                "       Enter  : 4 to apply pie chart"))
    if a==1:
        heatmap(data)
    elif a==2:
        scatter_plot(data)
    elif a==3:
        histogram_plot(data)
    elif a==4:
        pie_chart (data)
    else:
        break
print('Finally')
```
